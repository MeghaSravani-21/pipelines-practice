```python
import pandas as pd 
df1=pd.read_csv("C:\\Users\\Sravani\\Desktop\\us_customer_data.csv")
df2=pd.read_csv("C:\\Users\Sravani\\Desktop\\transaction_data.csv")
df=pd.merge(df1,df2, on="customer_id")
```

```python
df
```

```python
df1= df.drop_duplicates()
df2=df.drop_duplicates()
```

```python
#fixing the names column
df1['name'] = df1['name'].str.title()
df1['name'] = df1['name'].fillna('unknown')
```

```python
#dealing with emails column 
df1['email'] = df1['email'].fillna(df1['name'].str.replace(' ','').str.lower()+'@example.com')
```

```python
#fixing the amount with mean
df2['amount'] = df2['amount'].fillna(df2['amount'].mean())
```

```python
#extrcating country code from address
df1['countrycode']= df1['address'].str[-8:-6]
cc=df1['countrycode'].unique()
cc
```

```python
#convert datatype of phone column  to string 
df1['phone'] = df1['phone'].astype(str)
df1['phone'] = df1['phone'].str.replace('-','')
#dealing with the phone number 
def fix_value(df1):
    p=len(df1['phone'])
    for i in range(p):
        if len(df1['phone'][i])>10 or len(df1['phone'][i])<10:
            df1['phone'][i]='0'
       
    return df1
filtered_cvs=fix_value(df1)
```

```python
filtered_cvs.to_csv('filtered_csv.csv',index=False)
```
