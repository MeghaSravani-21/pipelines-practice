```python
orders=pd.read_csv("order_data.csv")
```

```python
orders.to_sql(name='orders',con=engine,if_exists='replace',index=False)
```

```python
read_orders=pd.read_sql_table('orders',con=engine)
```

```python
read_orders.columns
read_orders['order_status'].unique()
```

```python
query= ''' 
select 
customer_id,
count(order_id) as count,
sum(order_amount) as total_purchases
from 
orders 
where 
order_status='Completed' 
group by customer_id 
order by total_purchases desc
''' 
completed_orders=pd.read_sql_query(query,con=engine)
```
